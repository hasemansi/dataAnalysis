### Convert multi-dimensional arrays into one-dimensional arrays

```python
# M1) ravel :NumPy function (np.ravel() / arr.ravel()) : 
    # Returns a view of the original array. Changes to the new array may affect the original array and vice versa.

# M2) flatten :Array method (arr.flatten()) : 
    # Returns a copy of the original array. Changes to the new array do not affect the original array.

# M3) reshape / resize
```

```python
import numpy as np
```

```python
a=np.arange(1,21).reshape(5,4)
a
```

```python
a.ravel()
```

```python
np.ravel(a)
```

```python
a.flatten()
```

```python
#np.flatten(a)  # not work
```

#### Difference between ravel and flatten

```python
a
```

```python
ravelOp=a.ravel()
ravelOp
```

```python
flattenOp=a.flatten()
flattenOp
```

```python
ravelOp[1]=666
ravelOp
```

```python
a
```

```python
flattenOp[2]=333
flattenOp
```

```python
a
```

```python
# ravel and flatten method used to convert multi dimensional array to 1D array
# If you change any element either its output or original array it will reflect each other when we used ravel
# If you change any element either its output or original array it does not reflect each other when we used flatten. Changes reflected itself only
```

```python
a.reshape(20,)
```

```python
a.resize(20,)
```
