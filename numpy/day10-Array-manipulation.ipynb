### Masking / Boolean Indexing or Boolean Masking

```python
# Masking is a technique used in programming and data processing to select, hide, or modify data based on a condition.
# A mask is a boolean array (True/False) that controls which values are kept, removed, or changed.
```

```python
arr=np.arange(10)
arr
```

```python
arr>5
```

```python
arr[arr>5]
```

```python
arr[~(arr>5)]
```
