### Matrix operations

```python
# Element-wise operations:
# Addition: np.add()
# Subtraction: np.subtract()
# Division: np.divide()
# Multiplication: np.multiply()
```

```python
# Define two matrices to multiply
m1 = np.array([[1, 2], [3, 4]])
m2 = np.array([[5, 6], [7, 7]])
```

```python
m1
```

```python
m2
```

```python
np.add(m1,m2)
```

```python
np.subtract(m1,m2)
```

```python
np.divide(m1,m2)
```

```python
m1+m2
```

```python
m1-m2
```

```python
m1/m2
```

```python
m1*m2
```

```python
np.multiply(m1,m2)
```
