# Create array using np.arange() function
- Returns an array with evenly spaced elements as per the step size(by default 1)
- It has the following paramaters:
  * start : start of interval range. By default start = 0
  * stop  : end of interval range
  * step  : step size of interval. By default step size = 1
- Fractional step count is allowed in arange. Fractional stepsize not allowed in range.

```python
# Syntax: 
#  np.arange(start,      end,    stepsize,   dtype)
#           optional            optional   optional
#              0                   +1
                               # fractional step count is allowed 
```

#  Create array Using np.linspace() function
- Returns evenly spaced numbers over a specified interval. Here the start and stop both are included.            
- It has the following parameters:              
    - start: start of interval range. By default start = 0
    - stop: end of interval range
    - num : No. of samples to generate. By default num = 50

```python
#Syntax :- 
#np.linspace(start, stop,  num,    endpoint=True,    retstep=False)
#                         (part)   include/exclud    Gives stepsize
#                        optional    optional        optional
#                          50          True            False
```

```python
a=np.linspace(1,10)
a
```
