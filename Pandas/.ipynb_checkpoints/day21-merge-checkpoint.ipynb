```python
data=pd.merge(cat,Eng , on="Roll No", how="left")
data.head(5)
```

```python
Phy.head(5)
```

```python
data2=pd.merge(data,Phy , on=["First Name","Last Name"],how="left")
data2.head(5)
```

```python
Chem.head(5)
```

```python
data3=pd.merge(data2,Chem , left_on= ["First Name","Last Name"],right_on=["FName","LName"],how="left")
data3.head(5)
```

```python
data4=pd.merge(data3,Math,on=["First Name","Last Name"],how="left")
data4.head(5)
```

```python
data5=pd.merge(data3,Math,on=["First Name","Last Name"],how="left",suffixes=("_data3","_math"))
data5.head(5)
```

```python
data5=pd.merge(data3,Math,on=["First Name","Last Name"],how="right",suffixes=("_data3","_math"))
data5.head(5)
```

```python
data5=pd.merge(data3,Math,on=["First Name","Last Name"],how="inner",suffixes=("_data3","_math"))
data5
```
