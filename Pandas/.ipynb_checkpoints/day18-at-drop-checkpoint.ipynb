### Slicing data from the Pandas DataFrame:

```python
 # 1) loc for label-based slicing:Selecting rows and columns using labels # last index included
 # 2) iloc for positional slicing:Selecting rows and columns by integer position # last index excluded
 # 3) at for quick access to single values
 # 4) iat for quick access to single values by position
```

```python
import pandas as pd
```

```python
dt=pd.read_excel(r"C:\Users\hi\Documents\Desktop\Minskole classes\Pandas\inputData.xlsx")
dt
```

```python
dt.loc[0:5]
```

```python
dt.loc[0:5,["First Name","Last Name"]]
```

```python
dt.iloc[0:5]
```

```python
# dt.iloc[0:5,["First Name","Last Name"]]  ----> Not work
```

```python
# # Rename 
# # Syntax
# 1)Row labelled
#  dataframe_name.rename(index         = {1:"One",               4:"Four"} , inplace= True/False)
#                        For Row         Old value : New value,

# 2)Columns labelled
#  dataframe_name.rename(columns         = {1:"One",               4:"Four"} , inplace= True/False)
#                        For Column         Old value : New value,
```

```python
dt.rename(index={
    0:'Zero',
    1:'One',
    2:'Two'},
    inplace=True
    )
```

```python
dt
```

```python
dt.rename(columns={
    "English":"Eng",
    "Physics":"Phy"})
```

## Drop

```python
# # Drop # Syntax
# # 1) Drop record (row)
# dataframe_name.drop([row_index], inplace= True/False) # By default axis=0

# # 2) Drop column
# dataframe_name.drop([column_name], axis=1, inplace= True/False)
```

```python
# # If you want to delete
# rows use axis=0 ( By default axis value is 0 . Not mandadtory to define)
# Column use axis=1
```

```python
dt
```

```python
dt.drop("Two",inplace=True)
```

```python
dt
```

```python
dt.drop("Physics",axis=1)
```

```python
dt.drop([4,5])
```

```python
dt.drop(["English","Physics"],axis=1)
```

```python
dt
```

```python
dt.drop(columns=["Physics","Chemistry"])
```

```python
dt.drop(index=[3,4])
```
